```python
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt


plt.style.use("seaborn-darkgrid")
```

```python
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x, y)
```

Equação Normal

```python
x_b = np.c_[np.ones((100, 1)), x]
```

```python
theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
theta
```

```python
y_hat = theta[0] + x * theta[1]
```

```python
fig, ax = plt.subplots(figsize
                       =(15, 8))
plt.plot(x, y_hat)

plt.title('Gráfico usando Álgebra Linear')
plt.xlabel('Eixo x tendendo ao infinito')
plt.ylabel('Eixo y tendendo ao infinito')

plt.scatter(x, y)
```

GRADIENT DESCENT

```python
eta = 0.1
n_interactions = 10
m = 100

theta = np.random.randn(2, 1)
fig, ax = plt.subplots(figsize=(15, 8))
plt.scatter(x, y)

for interaction in range(n_interactions):
    gradients = 2/m * x_b.T.dot(x_b.dot(theta) - y)
    theta = theta - eta * gradients
    y_hat = theta[0] + x*theta[1]


    ax.plot(x, y_hat, alpha=0.3 + interaction/ n_interactions, color='red')
```
